```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating a 2X3 matrix
A = np.array([[1, 2, 3],
              [4, 5, 6]])
print("Matrix A:\n", A)
type(A)
```

```python
# Zero Matrix
np.zeros((10,10))
```

```python
# Matrix Shape and Reshaping
print("Shape of A:", A.shape)
A_reshaped = A.reshape(3,2)
print("Reshaped A:\n", A_reshaped)
```

```python
# Matrix addition (must have same dimensions)
A = np.array([[1, 2, 3],
              [4, 5, 6]])
B = np.array([[7, 8, 9],
              [10, 11, 12]])
C = A + B
D = A - B
print("A + B =\n", C)
print("A - B =\n", D)
```

```python
# Scalar Mulitplication
E = 2 * A
print("2 * A =\n", E)
```

```python
# Matrix Multiplication
a_ = np.array([1, 2, 3])
b_ = np.array([4, 5, 6])
np.dot(a_, b_)
```

```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])
F = np.array([[1, 2],
              [3, 4],
              [5, 6]])
G = np.matmul(A, F)
print(A.shape, F.shape)
print("A X F =\n", G)
```

```python
# Matrix Transpose
print(A)
print(A.T)
```

```python
# Identity Matrix
id_ = np.eye(8,8)
print(id_)
```

```python
# Matrix Inverse
M = np.array([[4, 7],
              [2, 6]])

M_inv = np.linalg.inv(M)
print("M: ", M)
print("Inverse of M:\n", M_inv)

# Verifying identity M * M_inv = I
np.matmul(M, M_inv) == np.eye(2,2)
```

```python
# This does not work as expected because Matrix Multiplication and Inversion
# involve floating point numbers, and hence tiny rounding errors occur
# Using np.allclose() instead checks that all the elements are approximately equal 
# within a tolerance. It 

identity_check = np.allclose(np.matmul(M, M_inv), np.eye(2,2))
print(identity_check)
```

```python
# Determinant of M
det_M = np.linalg.det(M)
print("Determinant of M: ", det_M)
```

```python
# Z-Score Normalization
X = np.array([10, 12, 14, 16, 18])
mean = X.mean()
std = X.std()

Z = (X - mean) / std
print(Z)
```

```python
# Plotting Standardized Feature Values

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# Loading data
iris = load_iris()
X = iris.data
feature_names = iris.feature_names

# Scale the data (standardization: mean=0, std=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Plot: Raw data vs Scaled data
# plt.figure(figsize=(12, 5))

# Plot 1: Raw data
plt.subplot(1, 2, 1)
for i in range(X.shape[1]):
    plt.plot(X[:, i], label = feature_names[i])
plt.title("Raw Feature Values")
plt.xlabel("Sample Index")
plt.ylabel("Original Scale")
plt.legend()

# Plot 2: Scaled data
plt.subplot(1, 2, 2)
for i in range(X_scaled.shape[1]):
    plt.plot(X_scaled[:, i], label = feature_names[i])
plt.title("Standardized Feature Values")
plt.xlabel("Sample Index")
plt.ylabel("Standard Scale (Z-score)")

plt.legend()

plt.show()
```
